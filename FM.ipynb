# Modulação em Frequência (FM)

> Simulação de sinal, portadora e onda modulada em frequêcia utilizando Python (Princípios de Telecomunicações I - UFRN)



Primeiramente, vamos importar as bibliotecas a serem utilizadas.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Agora, vamos definir os parâmetros do sinal e da portadora.

```python
# Parâmetros do sinal
amplitude_sinal = 1.0
frequencia_sinal = 5  # Frequência do sinal em Hz

# Parâmetros da portadora
amplitude_portadora = 0.5
frequencia_portadora = 25  # Frequência da portadora em Hz

# Parâmetros da modulação
indice_modulacao = 2.0  # Índice de modulação
```

É preciso criar vetores de tempo para o sinal da portadora.

```python
# Vetor de tempo
tempo = np.linspace(0, 1, num=1000)  # Tempo de 0 a 1 segundo, com 1000 amostras retiradas

# Sinal
sinal = amplitude_sinal * np.sin(2 * np.pi * frequencia_sinal * tempo)

# Portadora
portadora = amplitude_portadora * np.sin(2 * np.pi * frequencia_portadora * tempo)
```

Vamos modular o sinal de frequência:

```python
# Sinal modulado em frequência
sinal_modulado = amplitude_portadora * np.sin(2 *np.pi*frequencia_portadora*tempo + (indice_modulacao*np.cos(2*np.pi*frequencia_sinal*tempo)))
```

Por fim, vamos plotar os gráficos do sinal, da portadora e do sinal modulado:

```python
# Plotando os gráficos
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(tempo, sinal)
plt.title('Sinal de Frequência')
plt.xlabel('Tempo')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 2)
plt.plot(tempo, portadora)
plt.title('Portadora de Frequência')
plt.xlabel('Tempo')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 3)
plt.plot(tempo, sinal_modulado)
plt.title('Sinal Modulado em Frequência')
plt.xlabel('Tempo')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
